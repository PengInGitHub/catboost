```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Basic Plotting: plot

```python
# plot() method
# by default: line plot

# dataframe.plot() and pd.series.plot()
# a time series
ts = pd.Series(data=np.random.randn(1000), index=pd.date_range('1/1/2015', periods=1000))
ts = ts.cumsum()
ts.plot();
# to note if the index consits of date, gcf().autofmt_xdate() would be called
```

```python
# dataframe plot()
df = pd.DataFrame(data=np.random.randn(1000,4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()
plt.figure(); df.plot();
```

```python
# df: one column against another
df1 = pd.DataFrame(data=np.random.randn(1000,2), columns=list('BC')).cumsum()
df1['A'] = pd.Series(list(range(len(df))))
df1.plot(x='A', y='B')
```
